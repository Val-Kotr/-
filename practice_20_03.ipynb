```
class T1:
  n =10
  def total(self, N):
    self.total = int(self.n)+int(N)

class T2:
  def total(self, s):
    self.total = len(str(s))

t1=T1()
t2=T2()
t1.total(45)
t2.total(45)
print(t1.total)
print(t2.total)
```

```
class Base:
  def __init__(self, n):
    self.numb = n
  def out(self):
    print(self.numb)

class One(Base):
  def multi(self, m):
    #self.numb = 1
    self.numb *= m

class Two(Base):
  def inlist(self):
    self.inlist = list(str(self.numb))
  def out(self):
    i = 0 
    while i < len(self.inlist):
      print(self.inlist[i])
      i+= 1

obj1 = One(45)
obj2 = Two('abc')

obj1.multi(2)
obj1.out()

obj2.inlist()
obj2.out()
```

```
class Base:
  def __init__(self, N):
    self.numb = N
  def out(self):
    self.numb /= 2
    print (self.numb)
class Subclass (Base):
  def out(self):
    print ("\n----")
    Base.out(self)
    print("----\n")
i = 0
while i<10:
  if 4<i<7:
    obj = Subclass(i)
  else:
    obj = Base(i)
  i+=1
  obj.out()
```

Практическое задание

```
class fruits:
  def __init__(self, t, c):
    self.taste = t
    self.color = c
  def eat(self):
    print ('Eating fruits')


class Banan(fruits):
  def __init__(self, t, c, w):
    super().__init__(t, c)
    self.weight = w


  def eat(self):
    print ('Eating banana')
  
class Apple(fruits):
  def __init__(self, t, c, w):

    super().__init__(t, c)
    self.weight = w

  def eat(self):
    print ('Eating apple')

banan1 = Banan("sweet", "yellow", 150)
apple1 = Apple("juicy", "red", 200)

banan1.eat()
apple1.eat()
print("banan1 is", banan1.taste, banan1.color, banan1.weight)
print("apple1 is", apple1.taste, apple1.color, apple1.weight)
```
